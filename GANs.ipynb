```python
from sdv.datasets.local import load_csvs
from sdv.datasets.demo import download_demo

# This is the default folder name that the GOogle Colab notebook uses.
# Change this if you have your own folder with CSV files.
FOLDER_NAME = 'content/'

try:
  datasets = load_csvs(folder_name='/content/')
except ValueError:
  print('You have not uploaded any csv files.')
  
```

```python
datasets.keys()
```

```python
data = datasets['finalpreprocessed_detailed']
data=data[data['CVD']==1]
data['id']=data.index
```

```python
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata()
```

```python
metadata.detect_from_dataframe(data=data)
```

```python
for col in data.columns:
  metadata.update_column(
    column_name=col,
    sdtype='categorical')
metadata.update_column(
    column_name='id',
    sdtype='id')
metadata.update_column(
    column_name='AGEP_A',
    sdtype='numerical')
metadata.update_column(
    column_name='WEIGHTLBTC_A',
    sdtype='numerical')
metadata.set_primary_key(column_name='id')
```

```python
metadata
```

```python
metadata.validate()
```

```python
data.shape
```

```python
metadata.save_to_json(filepath='my_metadata_v1.json')
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata.load_from_json(
    filepath='my_metadata_v1.json')
```

```python
from sdv.single_table import CTGANSynthesizer

synthesizer = CTGANSynthesizer(
    metadata, # required
    enforce_rounding=False,
    epochs=100,
    verbose=True
)
```

```python
synthesizer.fit(data)
synthetic_data = synthesizer.sample(num_rows=10000)
```

```python
synthetic_data.to_csv('synthetic_dataGAN.csv')
```

```python
from sdv.single_table import GaussianCopulaSynthesizer

synthesizer = GaussianCopulaSynthesizer(metadata)
synthesizer.fit(data)
```

```python
synthetic_data_gau = synthesizer.sample(num_rows=10000)
synthetic_data_gau.to_csv('synthetic_dataGC.csv')
```
