```python
#Combining CVD data and getting one predictive column with CVD health status
#CHDEV_A, ANGEV_A, MIEV_A, STREV_A
CVD_points=[]
for index, row in data21.iterrows():
  temp=0
  if(row['CHDEV_A']==1):
    temp=1
  if(row['ANGEV_A']==1):
    temp=1
  if(row['MIEV_A']==1):
    temp=1 
  if(row['STREV_A']==1):
    temp=1 
  CVD_points.append(temp)
data21['CVD']=CVD_points
d=['CHDEV_A', 'ANGEV_A', 'MIEV_A', 'STREV_A']
data21=data21.drop(columns=d)
```

```python
# Calculate the number of heart disease cases for each occupation
heart_disease_counts21 = data21[data21['CVD'] == 1].groupby('EMDINDSTN1_A').size()

# Calculate the total number of people in each occupation
occupation_counts21 = data21.groupby('EMDINDSTN1_A').size()
occupation_counts21 = occupation_counts21[occupation_counts21 >= 100]

# Calculate the heart disease rate for each occupation
heart_disease_rates21 = (heart_disease_counts21 / occupation_counts21) * 100


# Sort heart disease rates
sorted_indrates21 = heart_disease_rates21.sort_values(ascending=False)

# Remove zero values from sorted_rates
sorted_indrates21 = sorted_indrates21[sorted_indrates21 > 0]
sorted_indrates21=sorted_indrates21.tail(25)

# Bar chart visualization
plt.figure(figsize=(10, 5))
bars = plt.bar(range(len(sorted_indrates21)), sorted_indrates21.values)
plt.xlabel('Industry Code')
plt.ylabel('CVD Rate (%)')
plt.title('CVD Rate by Industry')

# Set x-axis tick labels to occupation codes
plt.xticks(range(len(sorted_indrates21)), [reversed_dictind[i] for i in sorted_indrates21.index.astype(int)], rotation='vertical', fontsize=8)

# Display the graph
plt.show()
```
